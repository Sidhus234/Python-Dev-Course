```python
import pandas as pd
import numpy as np
import os
np.random.seed(12345)
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df_subset = df[['Name', 'Age', 'Wage', 'Value']]
df_subset.head()
```

```python
def value_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0

df_subset['Wage'] = df_subset['Wage'].str[1:]
df_subset['Wage'] = df_subset['Wage'].apply(value_to_float)
df_subset['Value'] = df_subset['Value'].str[1:]
df_subset['Value'] = df_subset['Value'].apply(value_to_float)
```

```python
df_subset.head()
```

```python
df_subset['difference'] = df_subset['Value'] - df_subset['Wage']
```

```python
df_subset.sort_values(by='difference', ascending=False, inplace=True)
```

```python
df_subset.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,10))
sns.scatterplot(df_subset['Wage'], df_subset['Value'])
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool
```

```python
TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(Wage,Value)", "(@Wage, @Value)"),
    ("Name", "@Name")]
)

p = figure(title="Soccer 2019", x_axis_label='Wage', y_axis_label='Value', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('Wage', 'Value', size=10, source=df_subset)
show(p)
```
